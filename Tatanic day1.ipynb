## 1.2 前置知识：pandas和numpy的基础知识 

#### Pandas 是一个专门用于数据分析的开源Python库。目前，所有使用python语言研究和分析数据集的专业人士，在做相关统计分析和决策时，Pandas 都是他们的基础工具。详见：https://geek-docs.com/pandas/pandas-tutorials/pandas-tutorial.html

##### Pandas包含以下三个数据结构：系列(Series)，数据帧(DataFrame)，面板(Panel)。Series	1	一维带标签的同质数组，大小不可变。Data Frames	2	通用的二维带标签，大小可变的表格结构，其中列的类型可能不同。Panel	3	通用的三维带标签，大小可变的数组。

#### 可以使用不同的输入来创建pandas DataFrame，如−列表 字典 系列 Numpy ndarrays 另一个DataFrame

#### NumPy是Python语言的一个扩充程序库。支持大量高级的维度数组与矩阵运算，此外也针对数组运算提供大量的数学函数库。Numpy内部解除了Python的PIL(全局解释器锁),运算效率极好,是大量机器学习框架的基础库!
